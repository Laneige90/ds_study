```python
import pandas as pd
```

```python
left = pd.DataFrame({

        'Key':['k0','k4','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']
})

right = pd.DataFrame({

        'Key':['k0','k1','k2','k3'],
        'C':['c0','c1','c2','c3'],
        'D':['d0','d1','d2','d3']
            
})
```

```python
left
```

```python
right
```

```python
# key를 기준으로 병합
pd.merge(left, right, on='Key')
```

```python
# left DataFrame을 기준으로 병합.  없는 부분은 NaN으로 출력
pd.merge(left, right, how='left', on='Key')
```

```python
# key를 기준으로 전부 병합.  없는 항목은 Nan으로 출력
pd.merge(left, right, how='outer', on='Key')
```

```python
# 공통분모만 병합
pd.merge(left, right, how='inner', on='Key')
```
