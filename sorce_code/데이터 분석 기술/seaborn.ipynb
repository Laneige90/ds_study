```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
# white 스타일
sns.set_style('white')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine()
plt.show()
```

```python
# whitegrid 스타일
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
# despine 적용
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(offset=10)
plt.show()
```

```python
# seaborn 기본 내장 데이터셋
tips = sns.load_dataset('tips')
tips.head()
```

```python
# boxplot 그리기
plt.figure(figsize=(8, 6))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
# boxplot에 칼럼 지정해 만들기
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# 칼럼 지정 및 구분
plt.figure(figsize=(8,6))
sns.boxplot(data = tips, x='day', y='total_bill', hue='smoker', palette='Set3')
plt.show()    
```

```python
# swarmplot
plt.figure(figsize=(8,6))
sns.swarmplot(x = 'day', y='total_bill', data=tips, color='.5')
plt.show()
```

```python
# 그래프 동시 적용
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x ='day', y='total_bill', data=tips, color='.23')
plt.show()
```

```python
# total_bill 과 tip 사이의 관계 파악
sns.set_style('darkgrid')
sns.lmplot(x ='total_bill', y='tip', data=tips)
plt.show()
```

```python
# lmplot > hue 옵션
sns.lmplot(x='total_bill', y='tip',hue='smoker' , data=tips)
plt.show()
```

```python
# flights 데이터셋
flights = sns.load_dataset('flights')
flights.head()
```

```python
# pivot 옵션 사용
flights = flights.pivot('month', 'year', 'passengers')
flights.head()
```

```python
# heatmap으로 경향 알아보기
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```

```python
# heatmap 색상 바꾸기
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
# iris 데이터셋
sns.set(style='ticks')
iris = sns.load_dataset('iris')
iris.head()
```

```python
# pairplot 다수의 칼럼 비교
sns.pairplot(iris)
plt.show()
```

```python
# pairplot의 hue 옵션
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# 원하는 칼럼만 pairplot
sns.pairplot(
    iris, x_vars=['sepal_width','sepal_length'], y_vars = ['petal_width', 'petal_length']
)
plt.show()
```
