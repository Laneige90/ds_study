## Pivot Table

```python
import pandas as pd
import numpy as np
```

### 판매 현황표 불러오기

```python
df = pd.read_excel('D:/data/02. sales-funnel.xlsx')
df.head()
```

### Name을 Index로 두고 재정렬

```python
# pivot table 사용
pd.pivot_table(df, index=['Name'])
```

### Index 여러개 지정 가능

```python
pd.pivot_table(df, index=['Name','Rep','Manager'])
```

```python
pd.pivot_table(df, index=['Manager','Rep'])
```

```python
# Values 지정해 출력
pd.pivot_table(df, index=['Manager','Rep'], values=['Price'])
```

```python
# 함수도 지정 가능
pd.pivot_table(df, index=['Manager','Rep'], values=['Price'], aggfunc=np.sum)
```

```python
# 갯수 적용
pd.pivot_table(df, index=['Manager','Rep'], values=['Price'], aggfunc=[np.mean, len])
```

```python
# 분류 지정
pd.pivot_table(
    df,
    index=['Manager','Rep'],
    values=['Price'],
    columns=['Product'],
    aggfunc=[np.sum]
)
```

```python
# NaN 처리 지정
pd.pivot_table(
    df,
    index=['Manager','Rep'],
    values=['Price'],
    columns=['Product'],
    aggfunc=[np.sum],
    fill_value= 0  # NaN에 원하는 숫자 넣기
)
```

```python
# 좀 더 깔끔하게 정리
pd.pivot_table(
    df,
    index=['Manager','Rep','Product'],
    values=['Price', 'Quantity'],
    aggfunc=[np.sum],
    fill_value=0
)
```
