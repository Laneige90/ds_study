## 인구 소멸 지역 파악

소멸 지역 시각화, 카트토그램 표현

### 데이터 읽기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
population = pd.read_excel('D://data/07_population_raw_data.xlsx', header=1)
population.fillna(method='pad', inplace=True)
population.head()
```

```python
# 칼럼 이름 변경
population.rename(
    columns={'행정구역(동읍면)별(1)':'광역시도', '행정구역(동읍면)별(2)':'시도', '계':'인구수'}, inplace=True
)
population.head()
```

```python
# 소계는 필요가 없다
population = population[(population['시도'] != '소계')]
```

```python
population.is_copy = False

population.rename(columns={'항목':'구분'}, inplace=True)

population.loc[population['구분'] == '총인구수 (명)', '구분'] = '합계'
population.loc[population['구분'] == '남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분'] = '여자'

population.head(10)
```

### 소멸지역 조사를 위한 데이터 생성

```python
population['20-39세'] = (
    population['20 - 24세']
    + population['25 - 29세']
    + population['30 - 34세']
    + population['35 - 39세']
)

population['65세 이상'] = (
    population['65 - 69세']
    + population['70 - 74세']
    + population['75 - 79세']
    + population['80 - 84세']
    + population['85 - 89세']
    + population['90 - 94세']
    + population['95 - 99세']
    + population['100+']
)

population.head(10)
```

```python
# pivot table로 데이터 정리
pop = pd.pivot_table(
    population, index=['광역시도','시도'],
    columns=['구분'],
    values=['인구수', '20-39세', '65세 이상']
)
pop
```

```python
# 소멸비율 계산  20-39세 여자 / (65세 이상 합계 / 2)
pop['소멸비율'] = pop['20-39세', '여자'] / (pop['65세 이상', '합계'] / 2)
pop.head()
```

```python
# 소멸비율로 소멸 위기 지역 체크도 가능
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
# index 재생성
pop.reset_index(inplace=True)
pop.head()
```

```python
# 칼럼 정리
tmp_columns = [
    pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n]
    for n in range(0, len(pop.columns.get_level_values(0)))
]
pop.columns = tmp_columns
pop.head()
```
